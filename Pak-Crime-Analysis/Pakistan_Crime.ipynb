***Robberies in Pakistan 2012 to 2017*** 

```python
roberry=df[df['Offence'] == 'Robbery']
roberry=roberry[['Year','Punjab','Sindh','KP','Balochistan','Islamabad']]
roberry=roberry.melt(id_vars=['Year'] , var_name='province' , value_name='Robbery Cases')
roberry=roberry.sort_values(by='Robbery Cases' , ascending=False)
roberry
```

```python
total_robbery_cases=df[df['Offence'] == 'Robbery'].agg({
    'Punjab':'sum',
    'Sindh':'sum',
    'KP':'sum',
    'Balochistan':'sum',
    'Islamabad':'sum'
}).reset_index()
total_robbery_cases.columns=['Province' , 'Total Robbery Cases']
total_robbery_cases=total_robbery_cases.sort_values(by='Total Robbery Cases' , ascending=False)
total_robbery_cases
```

***Visualization***

```python
total_robbery_cases.plot(
    x='Province', 
    y='Total Robbery Cases', 
    kind='barh', 
    figsize=(10, 6), 
    color='green',
    edgecolor='black',
    
)

# Add labels and title
plt.xlabel('Total Robbery Cases', fontweight='bold' , fontsize=10)
plt.ylabel('Province' , fontweight='bold' , fontsize=10)
plt.title('Total Robbery Cases by in Pakistan by Province( 2012 - 2017)' , fontweight='bold' , fontsize=12)
plt.show()
```

***Murder Cases in Pakistan 2012 to 2017***

```python
murders_cases=df[df['Offence'] == 'Murder']
murders_cases=murders_cases[['Year','Punjab','Sindh','KP','Balochistan','Islamabad']]
murders_cases=murders_cases.melt(id_vars=['Year'] , var_name='Province' , value_name='Murder Cases')
murders_cases=murders_cases.sort_values(by='Murder Cases' , ascending=False)
murders_cases
```

```python
Total_Murder_cases=df[df['Offence'] == 'Murder'].agg({
     'Punjab':'sum',
    'Sindh':'sum',
    'KP':'sum',
    'Balochistan':'sum',
    'Islamabad':'sum'
}).reset_index()
Total_Murder_cases.columns=['Province' , 'Total Murder Cases']
Total_Murder_cases=Total_Murder_cases.sort_values(by='Total Murder Cases' , ascending=False)
Total_Murder_cases['Years']=[2012,2015,2016,2017,2018]
Total_Murder_cases
```

***Visualization for Murders***

```python
plt.figure(figsize=(8,6))
barplot=sns.barplot(x='Total Murder Cases' , y='Province' , data=Total_Murder_cases , legend=False , hue='Province' ,edgecolor='grey')

barplot.set_xlabel('Total Murder Cases' , fontweight='bold' , fontsize=10)
barplot.set_ylabel('Province' ,fontweight='bold' , fontsize=10)


plt.title('Total Murder Cases in Pakistan by Province( 2012 - 2017)' ,fontweight='bold' , fontsize=12)
plt.show()
```
