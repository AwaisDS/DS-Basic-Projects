***Data Cleaning & PreProcessing***

```python
col_to_replace=['director' , 'cast' , 'country']
df[col_to_replace]=df[col_to_replace].replace(np.NaN , 'unknown')
df.head()
```

```python
col_to_drop=['date_added' , 'rating' , 'duration']
df.dropna(subset=col_to_drop , inplace=True)
df.sample(5)
```

```python
df['date_added'] = df['date_added'].astype('datetime64[ms]')
df.dtypes
```

```python
df['year_added']=df['date_added'].dt.year
print(df['year_added'])
```

```python
df['month_added'] = df['date_added'].dt.month_name()
df['month_added']
```

__Content Distribution(Movies Vs. TV Shows)__

```python
pd.crosstab(df.type , df.year_added)
```

```python
pd.crosstab(df.type , df.year_added).plot(kind='bar')
plt.tight_layout()
plt.xlabel("Content Type" , fontweight='bold' , fontsize=12)
plt.legend(title='Year Added', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.2)
plt.xticks(rotation=2)
plt.title("Netflix Content by Type and Year Added" , fontweight='bold' , fontsize=12)
plt.show()
```

```python
df.groupby(['year_added' , 'type']).size().reset_index(name='count').sort_values(by='count' , ascending=False).head(5)
```

```python
year_type_count = df.groupby(['year_added', 'type']).size().reset_index(name='count')
sns.lineplot(data=year_type_count, x='year_added', y='count', hue='type',palette='Set2')
plt.title("Content Added Over the Years by Type")
plt.xlabel("Year Added")
plt.grid(True)
plt.show()
```

```python
plt.tight_layout()
sns.histplot(data=df, x="year_added", bins=20, kde=True, color="darkcyan")
plt.title("Distribution of Netflix Content by Year Added")
plt.xlabel("Year Added")
plt.xticks(rotation=45)
plt.grid(True, which='both', axis='y', linestyle='-.', linewidth=0.7)
plt.show()
```

__Distribution of different content type on ***Netflix***__ 

```python
num_movies=df['type'].value_counts()['Movie']
num_tvshows=df['type'].value_counts()['TV Show']
print(f"Number of Movies: {num_movies}")
print(f"Number of TV Shows: {num_tvshows}")
```

```python
plt.tight_layout()
sns.countplot(data=df, x="type", color='tomato')
plt.title("Distribution of Content Types")
plt.xlabel("Content Type")
for index, value in enumerate(df['type'].value_counts()):
    plt.text(index, value, str(value), ha='center', va='bottom')
plt.show()
```

```python
plt.tight_layout()
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140,colors=['green', 'skyblue'])
plt.title('Percentage Distribution of Content Types')
plt.ylabel('')
plt.show()
```

__Countries which have the most content available on Netflix__

```python
top_10_countries = df['country'].replace("unknown", pd.NA).dropna().value_counts().head(10)
top_10_countries
```

```python
colors = sns.color_palette("Blues", len(top_10_countries))[::-1]
top_10_countries.plot(kind='barh', color=colors)
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Country")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
temp = df.groupby(['year_added', 'country']).size().reset_index(name='count').sort_values("count",ascending=False)
temp
```

```python
country_year_trends = df.groupby(["year_added", "country"]).size().reset_index(name='count')
top_countries = country_year_trends[country_year_trends['country'].isin(['United States', 'India', 'United Kingdom', 'Japan', 'South Korea'])]

sns.lineplot(data=top_countries, x='year_added', y='count', hue='country')
plt.title("Content Distribution by Country Over the Years")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.tight_layout()
plt.show()
```

 __Most popular genres available on Netflix__

```python
top_10_genres_count = df['listed_in'].str.split(', ').explode().value_counts().head(10)
colors = sns.color_palette("Purples", len(top_10_genres_count))[::-1]
plt.figure(figsize=(10,5))
top_10_genres_count.plot(kind='bar', color=colors)
plt.title("Top 10 Most Popular Genres on Netflix", fontsize=16)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

__Finding the average duration of Movies__

```python
movie_data = df[df['type'] == "Movie"]
avg_movie_duration = movie_data['duration'].str.replace(' min', '').astype(int).mean()
print(f"Average Movie Duration : {avg_movie_duration:.2f} minutes")
```

 __Duration of movies distributed on Netflix__

```python
movie_duration = df[df['type']=='Movie']['duration'].str.replace(' min','').astype(int)
sns.histplot(movie_duration, bins=20, kde=True, color='skyblue')
plt.title("Distribution of Movie Duration on Netflix")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

```python
sns.boxplot(y=movie_duration, color='green')
plt.title("Box Plot for Movie Duration")
plt.ylabel("Duration (Minutes)")
plt.tight_layout()
plt.show()
```

__Number of content additions vary by month throughout the year 2021__

```python
df['month_added'].value_counts().reindex(pd.date_range('2021-01', '2021-12',freq='ME').strftime('%B'), fill_value=0)
```

```python
month_no = df['date_added'].dt.month
heatmap_data = pd.crosstab(df['year_added'], month_no)
sns.heatmap(data=heatmap_data, cmap='Greens', linewidths=.5)
plt.title("Heatmap of Content Added by Year and Month")
plt.xlabel("Month")
plt.ylabel("Year")
plt.tight_layout()
plt.show()
```

***Netflix Dataset Analysis Summary***

__Dataset Overview:__

The dataset contains information about Netflix content, including movies and TV shows.
Key columns: title, type (Movie/TV Show), director, cast, country, date_added, release_year, rating, duration, and listed_in.

__Objective:__

The aim of the analysis is to understand the distribution of content types, the growth of content over the years, content ratings, and other key insights.

__Content Distribution:__

Type of Content: Analysis of the ratio between Movies and TV Shows. For example, there might be more movies compared to TV shows.
Distribution by Genre: Analyzed which genres are most common in Netflix content. The genres could include drama, comedy, thrillers, and more.

__Year-wise Addition of Content:__

Trend Over Time: Analyzed how the number of shows and movies added to Netflix has changed over the years, with a focus on recent years.

__Visualization:__ A histogram showed the distribution of content by year_added, revealing trends like a significant increase in content additions in the past few years.

__Ratings Distribution:__

Investigated the different types of content ratings such as TV-MA, TV-14, R, etc., and their distribution across the content.
Visualized the ratings using bar plots to see which ratings are more common.

__Duration Analysis:__

Movies: Examined the duration of movies to see what the average or most common length is.
TV Shows: Analyzed the number of seasons of TV shows and identified any patterns in the number of episodes or seasons.

__Country-wise Analysis:__

Analyzed which countries contribute the most content to Netflix. For example, the U.S., India, and the UK might be the leading contributors.
This was visualized using bar charts showing the top 10 countries in terms of content contribution.

__Top Directors and Actors:__

Analyzed which directors and actors appear most frequently in Netflix content. This helps identify popular creators on the platform.

__Key Visualizations:__

Histogram of content distribution by the year it was added.
Bar plot of content types (Movies vs. TV Shows).
Country-wise content distribution.
Rating distribution and its comparison across content types.
Genre distribution visualization.


***Conclusions:***

__Content Growth:__
 Netflix has significantly ramped up the addition of new content, especially after 2015.

__Movie Dominance:__
 Movies likely outnumber TV shows on the platform.
 
__Diverse Content:__ 
Netflix offers content from a wide variety of genres and countries, catering to a global audience.
